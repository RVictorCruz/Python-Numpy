## **<font color=#2dd8f6>Fórmula para calcular o coeficiente angular e linear da reta</font>**

```
A = Coneficiente Angular
n -> número de elementos
Y= Moscow
X = datas

O coeficiente angular pode ser obtido usando a equação:


```

image.png

``````python
Y = Moscow
X = datas
n = np.size(Moscow)
``````

``````python
a = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y))/(n*np.sum(X**2) - np.sum(X)**2)
``````

## **<font color=#2dd8f6>Calculando o Coeficiente linear</font>**

image.png

``````python
b = np.mean(Y) - a * np.mean(X)
``````

## **<font color= #92f5fd>Equação da reta</font>**

```
y = ax + b
```

``````python
y = a * X +b
``````

``````python
np.linalg.norm(Moscow - y)
``````

``````python
plt.plot(datas, Moscow)
plt.plot(X,y)
plt.plot(24, 24*a+b, "*r")
plt.plot(100, 100*a+b, "*r")
``````

### **Números aleatórios com a Numpy**

### Testando o coeficiente angular

```
Inserindo as informações dentro de um array
```

``````python
norma = np.array([])
for i in range(100):
  norma = np.append(norma, np.linalg.norm(Moscow-(coef_angulares[i]*X+b)))
``````
